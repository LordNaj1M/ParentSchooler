```python
import pandas as pd

data = pd.read_csv('journal_scrape_with_keywords.csv')
```

```python
data
```

```python
import langdetect
from langdetect import detect
```

```python
def detect_language(abstract) :
    language = detect(abstract)
    
    if language == 'id' :
        language = 'Indonesia'
    elif language == 'en' :
        language = 'English'
    else :
        language = 'Others'
        
    return language
```

```python
data['language'] = data['abstract'].apply(lambda abstract : detect_language(str(abstract)))
```

```python
data.to_csv('complete_journal_data.csv', index = False)
```
